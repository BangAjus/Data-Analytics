```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('WineQT.csv')
df
```

```python
df.info()
```

```python
q = df['quality'].values
q = np.unique(q)
q
```

```python
am = df['quality'].value_counts().reset_index().sort_values(['index'])
am = [i for i in am['quality']]
am[1:]
```

```python
df['quality'].value_counts().reset_index().sort_values(['index'])
```

```python
df.corr()['quality'].sort_values(ascending=False)
```

```python
test = df[df['quality'] == 3][-2:]
for i in range(4, 9):
    j = 1
    dfx = df[df['quality'] == i][-1*(am[j]//3):]
    test = pd.concat([test, dfx])
    j += 1    
nott = ['alcohol', 'sulphates', 'citric acid', 'fixed acidity', 'volatile acidity', 'total sulfur dioxide', 'quality']
for i in test.columns:
    if i not in nott:
        test = test.drop(columns=[i])
x_test = test.drop(columns=['quality'])
y_test = test[['quality']]
```

```python
train = df.copy()
col2 = ['alcohol', 'sulphates', 'citric acid', 'fixed acidity', 'volatile acidity', 'total sulfur dioxide', 'quality']
for i in train.columns:
    if i not in col2:
        train = train.drop(columns=[i])
x_train = train.drop(columns=['quality'])
y_train = train[['quality']]      
```
