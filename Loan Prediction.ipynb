```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tr = pd.read_csv('loan_sanction_train.csv')
ts = pd.read_csv('loan_sanction_test.csv')
```

# Clean the train data

```python
tr.nunique()
```

```python
for i in tr.columns:
  print(tr[i].isna().value_counts())
```

```python
tr['Loan_Amount_Term'].value_counts()
```

```python
# to avoid more data loss, i have to fill the numeric data with mean
for i in tr.columns:
  if tr[i].dtypes != 'object':
    tr[i] = tr[i].fillna(tr[i].mean())

for i in ts.columns:
  if ts[i].dtypes != 'object':
    ts[i] = ts[i].fillna(ts[i].mean())
```

```python
tr = tr.dropna()
ts = ts.dropna()
```

```python
tr.info()
```

#Prepare the data

```python
tr.nunique()
```

```python
tr.info()
```

```python
for i in tr.columns:
  if tr[i].dtypes == 'object':
    print(tr[i].value_counts())
```

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
for i in tr.columns:
  if i != 'Loan_ID':
    if tr[i].dtypes == 'object':
      tr[i] = lb.fit_transform(tr[i])

for i in ts.columns:
  if i != 'Loan_ID':
    if ts[i].dtypes == 'object':
      ts[i] = lb.fit_transform(ts[i])
```

```python
X, y = tr.drop(columns=['Loan_ID', 'Loan_Status']), tr['Loan_Status']
```

# Train the data

```python
from autosklearn.classification import AutoSklearnClassifier
import sklearn.model_selection
import sklearn.metrics
```

```python
model = AutoSklearnClassifier(time_left_for_this_task=60)
model.fit(X, y)
```

```python
model.leaderboard(detailed=True)
```

```python
X_tst, y_tst = X[-50:], y[-50:]
```

```python
model.score(X_tst, y_tst)
```

```python
Xt = ts.drop(columns=['Loan_ID'])
ts['Loan_Status'] = model.predict(Xt)
```
